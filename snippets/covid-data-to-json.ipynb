```
import pandas as pd
import json
```

```
url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/03-26-2020.csv"
df = pd.read_csv(url)
```

```
df.head()
```

```
data = df.loc[df["Country_Region"] == "US"]
```

```
data = data.drop(columns=["FIPS", "Country_Region", "Last_Update", "Lat", "Long_", "Combined_Key"])
data.rename(columns = {'Admin2':'County', 'Province_State':'State'}, inplace = True)
data.head()
```

```
data_dict = data.to_dict(orient='records')
print(json.dumps(data_dict, indent=2))
```
